```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
L = 0.08
M = 0.03
uS = M*343
Bm = np.arange(5,50,1)
t = 0.001 
T = 0.001*uS/L
k = np.arange(0.0001,0.0002,0.00001)
Bt = []

for j in Bm: # wothout t, Bt(Bm,k)
    for i in k:
        bt = j*(L/uS*math.pi/i*math.cos(L/uS*math.pi/i*t)-1/T*math.sin(L/uS*math.pi/i*t)*math.exp(-t/T))
        Bt.append(bt) 

    
plt.plot(Bt)
```

```python
B = []
for j in Bm:
    for i in k:
        b = j * math.sin(math.pi * t / i) * math.exp(-t/T)
        B.append(b) 

from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection ='3d')

k1 = np.linspace(0.0001,0.0002,len(B))
Bm1 = np.linspace(5,50,len(B))
ax.plot3D(B, k1, Bm1, 'green')
ax.set_xlabel("B")
ax.set_ylabel("k")
ax.set_zlabel("Bm", labelpad=-30)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection ='3d')
Bt = []
t = 0
for j in Bm: # Bt(Bm,k,t)
    for i in k:
        while t < T:
            bt = j*(L/uS*math.pi/i*math.cos(L/uS*math.pi/i*t)-1/T*math.sin(L/uS*math.pi/i*t)*math.exp(-t/T))
            Bt.append(bt)
            t += 0.01 
    t = 0

t1 = np.linspace(0,T,len(Bt))
Bm1 = np.linspace(5,50,len(Bt))
ax.plot3D(Bt, t1, Bm1, 'green')
ax.set_xlabel("Bt")
ax.set_ylabel("t")
ax.set_zlabel("Bm", labelpad=-30)
```
